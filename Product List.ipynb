```python
from demyst.analytics import Analytics
import pandas as pd

analytics = Analytics()
```

```python
# Download product data
product_data = analytics.products()

# Format into Product Name, Category, Tags, Website, and Description
main_components = list(map(lambda product: {'Product Name': product['name'], 'Category' : product['category'], 'Tags' : ", ".join(list(map(lambda tag: tag["name"], product['tags']))), "Website" : product["website"], "Description" : product["description"]}, product_data))

# Read in as DataFrame
products = pd.DataFrame(main_components)
products
```
